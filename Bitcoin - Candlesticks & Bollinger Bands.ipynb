```python
from math import pi

import pandas as pd

from bokeh.plotting import figure, show, output_file, output_notebook
from bokeh.sampledata.stocks import MSFT
```

```python
output_notebook()
```

```python
df = pd.DataFrame(MSFT)[:50]
df["date"] = pd.to_datetime(df["date"])

inc = df.close > df.open #increase - get all rows for which close price was higher than open price
dec = df.open > df.close #decrease - get all rows for which close price was lower than open price
w = 12*60*60*1000 # half day in ms

TOOLS = "pan,wheel_zoom,box_zoom,reset,save"

p = figure(x_axis_type="datetime", tools=TOOLS, plot_width=900, title = "MSFT Candlestick")
p.xaxis.major_label_orientation = pi/4 #changes X-label angle to pi/4 = 45 degrees
p.grid.grid_line_alpha=0.3

p.segment(df.date, df.high, df.date, df.low, color="black")
p.vbar(df.date[inc], w, df.open[inc], df.close[inc], fill_color="#088A08", line_color="black")
p.vbar(df.date[dec], w, df.open[dec], df.close[dec], fill_color="#B40404", line_color="black")

output_file("candlestick.html", title="candlestick.py example")
```

```python
show(p)  # open a browser
```

```python
def bollinger(df_prices, window):
    
    df_prices['ma'] = df_prices.loc[:]['Close'].rolling(window).mean()
    
    df_prices['std'] = df_prices.loc[:]['Close'].rolling(window).std()
    
    df_prices['Upper Band'] = df_prices.loc[:]['ma'] + (df_prices.loc[:]['std']*2)
    
    df_prices['Lower Band'] = df_prices.loc[:]['ma'] - (df_prices.loc[:]['std']*2)
    
    return df_prices
```

```python
btc_bollinger = bollinger(mydata, 20)
```

```python
inc = mydata.Close > mydata.Open #increase - get all rows for which close price was higher than open price
dec = mydata.Open > mydata.Close #decrease - get all rows for which close price was lower than open price
w = 12*60*60*1000 # half day in ms

TOOLS = "pan,wheel_zoom,box_zoom,reset,save"

p = figure(x_axis_type="datetime", tools=TOOLS, plot_width=900, title = "BTC Candlestick")
p.xaxis.major_label_orientation = pi/4 #changes X-label angle to pi/4 = 45 degrees
p.grid.grid_line_alpha=0.3

p.segment(mydata.index, mydata.High, mydata.index, mydata.Low, color="black")
p.vbar(mydata.index[inc], w, mydata.Open[inc], mydata.Close[inc], fill_color="#088A08", line_color="black")
p.vbar(mydata.index[dec], w, mydata.Open[dec], mydata.Close[dec], fill_color="#B40404", line_color="black")

#Plot Bollinger Bands

#Lower Band
p.line(btc_bollinger.index, btc_bollinger['Lower Band'], line_width=1, line_color='orange')

#Lower Band
p.line(btc_bollinger.index, btc_bollinger['Upper Band'], line_width=1, line_color='blue')

output_file("candlestick.html", title="candlestick.py example")
```
